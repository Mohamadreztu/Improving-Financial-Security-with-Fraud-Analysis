```python
import pandas as pd
```

```python
data_conv = pd.read_csv('../dataset/transaction_online_payments.csv')
data_conv
```

```python
# Menghapus kolom type, nameOrig, dan nameDest
data_conv = data_conv.drop(columns=['type', 'nameOrig', 'nameDest'])
```

```python
col = data_conv.columns.to_list()
col
```

```python
# Menggunakan apply untuk mengubah nilai kolom isFraud
data_conv['isFraud'] = data_conv['isFraud'].apply(lambda x: 'Tidak Penipuan' if x == 0 else 'Penipuan')

data_conv.head()
```

```python
step=[data_conv['step'].to_list()]
amount=[data_conv['amount'].to_list()]
oldbalanceOrg=[data_conv['oldbalanceOrg'].to_list()]
newbalanceOrig=[data_conv['newbalanceOrig'].to_list()]
oldbalanceDest=[data_conv['oldbalanceDest'].to_list()]
newbalanceDest=[data_conv['newbalanceDest'].to_list()]
isFraud=[data_conv['isFraud'].to_list()]
isFlaggedFraud=[data_conv['isFlaggedFraud'].to_list()]
step
```

```python
# Jadikan Sebagai Dictionary
dic = {
    col[0]: step,
    col[1]: amount,
    col[2]: oldbalanceOrg,
    col[3]: newbalanceOrig,
    col[4]: oldbalanceDest,
    col[5]: newbalanceDest,
    col[6]: isFraud,
    col[7]: isFlaggedFraud
}
```

```python
data_conv = pd.DataFrame(dic)
data_conv
```

```python
# Simpan DataFrame ke file Excel
data_conv.to_excel('../visualization/transaction_online_payments.xlsx', index=False, engine='openpyxl')
```
